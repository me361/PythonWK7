```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set Seaborn style
sns.set(style="whitegrid")

# Show plots inside the notebook
%matplotlib inline
```

```python
# Load the iris dataset from CSV
df = pd.read_csv('iris_dataset.csv')

# Display the first few rows
df.head()
```

```python
# Check the structure of the dataset
df.info()

# Check for any missing values
df.isnull().sum()
```

```python
# Basic statistics for numerical columns
df.describe()
```

```python
# Group by species and get the mean of each feature
df.groupby('species').mean()
```

```python
# Map numeric species values to actual names
species_map = {0: 'setosa', 1: 'versicolor', 2: 'virginica'}
df['species'] = df['species'].map(species_map)

# Let's check if it worked
df.head()
```

```python
df.groupby('species').mean ()
```

```python
print(df.columns.tolist())
```

```python
df['petal length (cm)'].head()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df.index, df['petal length (cm)'], label='Petal Length (cm)', color='green')
plt.title('Trend of Petal Length Across Dataset')
plt.xlabel('Index')
plt.ylabel('Petal Length (cm)')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
df.groupby('species')['petal length (cm)'].mean().plot(kind='bar', color='skyblue')
plt.title('Average Petal Length by Species')
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df['sepal width (cm)'], bins=20, color='purple', edgecolor='black')
plt.title('Distribution of Sepal Width')
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='sepal length (cm)', y='petal length (cm)', hue='species')
plt.title('Sepal Length vs Petal Length by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.legend(title='Species')
plt.show()
```
